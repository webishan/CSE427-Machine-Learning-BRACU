# Numpy

# 1.Create two linear arrays a and b with 20 random elements each ranging from 1 to 100.

```python
import numpy as np
```

```python
a = np.random.randint(1, 101, size=20)
b = np.random.randint(1, 101, size=20)

print("Array a:", a)
print("Array b:", b)
```

# 2. Now change those linear arrays to two 5x4 matrices.

```python
a = a.reshape(5, 4)
b = b.reshape(5, 4)

print("Matrix a:\n", a)
print("\nMatrix b:\n", b)
```

# 3. Perform scalar multiplication with matrix a and b.

```python
scalar = a * b
print(scalar)
```

# 4. Now try to modify the matrix b and perform matrix multiplication with a and b to store it in c

```python
b_T = b.T

c = a @ b_T

print("Result matrix c:\n", c)
```

# 5. Print only the column 2, 3 of matrix c. Print the whole matrix using slicing technique.

```python
print("\nColumns 2 and 3 of matrix c:\n", c[:, 1:3])

print("Whole matrix c (using slicing):\n", c[:, :])
```

# 6. Find the largest element in the array, and also the index of that element.

```python
max_value = np.max(c)

max_index = np.unravel_index(np.argmax(c), c.shape)

print("Largest element:", max_value)
print("Index of largest element (row, column):", max_index)
```

# 7. Convert the matrix c into an one dimensional array keeping the original array unchanged.

```python
c_flat = c.flatten()

print("Original matrix c:\n", c)
print("\nFlattened 1D array:\n", c_flat)
```

# Matplotlib

# 1. Generate a scatter plot graph for temperature_dhaka vs humidity_dhaka using (*) as the marker in color red.

```python
import matplotlib.pyplot as plt
```

```python
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]

plt.figure(figsize=(8, 5))
plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')
plt.title('Scatter Plot: Temperature vs Humidity in Dhaka')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.grid(True)
plt.show()
```

# 2. Now do a line plot for study_hours vs marks with figure size (12,8)

```python
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

plt.figure(figsize=(12, 8))
plt.plot(study_hours, marks, marker='o', linestyle='-', color='blue')
plt.title('Line Plot: Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

# 3. Now draw a horizontal bar graph showing the scores in each of the subjects.

```python
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]

plt.figure(figsize=(8, 5))
plt.barh(subjects, marks, color='skyblue')
plt.title('Scores in Each Subject')
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# 4. Draw a vertical bar graph where each subject score is visualized in different colors.

```python
colors = ['red', 'blue', 'green', 'orange', 'purple']

plt.figure(figsize=(8, 5))
plt.bar(subjects, marks, color=colors)
plt.title('Vertical Bar Graph: Subject Scores in Different Colors')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 5. Show all these graphs in one graph using subplot( ) function.

```python

study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks_study = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks_subjects = [89, 90, 45, 78, 99]
colors = ['red', 'blue', 'green', 'orange', 'purple']

temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]

plt.figure(figsize=(16, 12))

plt.subplot(2, 2, 1)
plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')
plt.title('Temperature vs Humidity (Dhaka)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(study_hours, marks_study, marker='o', linestyle='-', color='blue')
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.barh(subjects, marks_subjects, color='skyblue')
plt.title('Scores in Each Subject (Horizontal)')
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.subplot(2, 2, 4)
plt.bar(subjects, marks_subjects, color=colors)
plt.title('Scores in Each Subject (Vertical)')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
